

**Project Goal**: Gathering and analyzing information from unstructured sources for further analysis and data output in a standardized catalog format.

**Example Task**: For the example, I will take a scenario that will collect the assortment of motherboards for building computer systems for tasks such as gaming, programming, and artificial intelligence. I chose this example because I have an understanding of the subject matter.

**Let's Get Started**
Formalization:
- I am collecting a product catalog from different suppliers and providing a structured table.

The system skeleton consists of 3 major classes, which are the foundation for gathering and structuring information:
- Entity(class ***Product***)
- Supplier(class ***Supplier***)
- Web driver(class ***Driver***)

Let's start with the web driver: its task is to communicate with the browser. By default, I have chosen Mozilla since it is commonly used for such solutions. However, other browsers can also be used, although I haven't tested them yet: ).
