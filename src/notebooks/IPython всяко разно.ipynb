```python
from IPython.display import clear_output
import ipywidgets as widgets

# Создаем кнопку для запуска
button = widgets.Button(description="Запустить")

# Функция, которая будет выполняться при нажатии кнопки
def on_button_click(b):
    os.system("python main.py")
    print("Запуск выполнен успешно!")

# Привязываем функцию к событию нажатия кнопки
button.on_click(on_button_click)

# Создаем вывод без поля ввода
output = widgets.Output(layout={'border': '1px solid black'})
output.append_display_data(button)

# Очищаем вывод
clear_output(wait=True)

# Отображаем вывод без поля ввода
display(output)
```
