```python
from urllib.parse import urlparse, parse_qs
```

```python
parsed_url = urlparse(url)._asdict()
```

```python
print(parsed_url)
```

```python
for key in parsed_url._asdict():
    print(key)
```

```python
print(type(parsed_url))
```

```python
query_params = parse_qs(parsed_url.query)
for key,value in query_params.items():
    print(key,value)
```
